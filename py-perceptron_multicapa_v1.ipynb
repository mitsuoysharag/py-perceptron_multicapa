```python
def nonlin(x,deriv=False):
    if(deriv==True):
        return x*(1-x)
    return 1/(1+np.exp(-x))

np.random.seed(1)
syn0 = 2 * np.random.random((len(X[0]),30)) - 1
syn1 = 2 * np.random.random((30,20)) - 1
syn2 = 2 * np.random.random((20,len(Y[0]))) - 1

for j in range(10000):
    l0 = X
    l1 = nonlin(np.dot(l0,syn0))
    l2 = nonlin(np.dot(l1,syn1))
    l3 = nonlin(np.dot(l2,syn2))

    l3_error = Y - l3
    if (j% 1000) == 0:
        print ("Error: " + str(np.mean(np.abs(l3_error))))
    l3_delta = (l3_error) * nonlin(l3,deriv=True)
    l2_delta = (l3_delta.dot(syn2.T)) * nonlin(l2,deriv=True)
    l1_delta = (l2_delta.dot(syn1.T)) * nonlin(l1,deriv=True)
    
    syn2 += l2.T.dot(l3_delta)
    syn1 += l1.T.dot(l2_delta)
    syn0 += l0.T.dot(l1_delta)
```
